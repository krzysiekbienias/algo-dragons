# House robber
### Problem description
### Mathematic Formula
$$
rob(i) =
\begin{cases} 
0, & i \geq \text{len}(arr) \\ 
\max(arr[i] + rob(i+2), rob(i+1)), & i < \text{len}(arr)
\end{cases}
$$
### Time and space complexity
### Similar challange.
Please not that similar is `max_subset_sum_no_adjacent`
### Code 

```python
def house_robbery(arr, i=0, lookup=None):
    """
    Calculate the maximum amount of money that can be robbed from a series of houses 
    without robbing two adjacent houses.

    This function uses recursion with memoization to determine the optimal amount of money that can be robbed 
    from a list of houses represented by an integer array. Each house has a certain amount of money, 
    and the function ensures that no two directly adjacent houses are robbed.

    Parameters
    ----------
    arr : list of int
        A list of integers representing the amount of money at each house.
    i : int, optional
        The current index in the list being considered (default is 0).
    lookup : dict, optional
        A dictionary used for memoization to store results of previously calculated indices (default is None).

    Returns
    -------
    int
        The maximum amount of money that can be robbed without robbing two adjacent houses.

    Examples
    --------
    >>> house_robbery([2, 7, 9, 3, 1])
    12  # (2 + 9 + 1)

    >>> house_robbery([1, 2, 3, 1])
    4   # (1 + 3)

    >>> house_robbery([5, 3, 4, 11, 2])
    16  # (5 + 11)
    """
    lookup = {} if lookup is None else lookup
    if i in lookup:
        return lookup[i]
    if i >= len(arr):
        return 0
    else:
        lookup[i] = max(arr[i] + house_robbery(arr, i + 2, lookup), house_robbery(arr, i + 1, lookup))
        return lookup[i]

```

```python
def max_subset_sum_no_adjacent(array,i=0,lookup=None):
    """
    Calculate the maximum subset sum from an array without selecting adjacent elements.

    This function uses recursion with memoization to find the maximum sum of non-adjacent
    elements in the given list. The function iterates through the array and at each index determines
    whether to include the current element or skip it, ensuring that adjacent elements are not selected.

    Parameters
    ----------
    array : list of int
        A list of integers for which the maximum subset sum is to be calculated.
    i : int, optional
        The current index in the array being considered (default is 0).
    lookup : dict, optional
        A dictionary used for memoization to store results of previously calculated indices (default is None).

    Returns
    -------
    int
        The maximum sum of non-adjacent elements in the array.

    Examples
    --------
    >>> max_subset_sum_no_adjacent([3, 2, 5, 10, 7])
    15  # (3 + 10 + 2)
    
    >>> max_subset_sum_no_adjacent([1, 2, 3, 4])
    6  # (1 + 3 or 2 + 4)

    >>> max_subset_sum_no_adjacent([5, 1, 1, 5])
    10  # (5 + 5)
    """
    # the same challange like house robber
    lookup={} if lookup is None else lookup
    # base case
    if i>=len(array):
        return 0
    if i in lookup:
        return lookup[i]
    else:
        lookup[i]=max(array[i]+max_subset_sum_no_adjacent(array,i+2,lookup),max_subset_sum_no_adjacent(array,i+1,lookup))
        return lookup[i]
```
